```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
#
url= "https://www.worldometers.info/coronavirus/#countries"
data=requests.get(url).text
```

```python
soup=BeautifulSoup(data,'lxml')
```

```python
get_data=soup.find("table",id="main_table_countries_today")
get_table_data=get_data.tbody.find_all('tr')
```

```python
dict={}
for i in range(len(get_table_data)):
    try:
        key=get_table_data[i].find_all('a',href=True)[0].string
    except:
        key=get_table_data[i].find_all('td')[0].string
        
    values=[j.string for j in get_table_data[i].find_all('td') ]
    
    dict[key]=values
print(dict)
    
```

```python
col_names=["Total cases","New cases","total Deaths","Total Recoverd","Active","Serious Critical","Total cases/1M pop","Total deaths/1M pop"]
df=pd.DataFrame(dict).iloc[1:,:].T
del df[1]
df
df1=df.iloc[:,:8]
df1
df1.index_name="Country"
df1.columns=col_names
df1
```
