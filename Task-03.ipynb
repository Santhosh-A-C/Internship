```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import plotly.express as px
```

```python
import plotly.io as pio
```

```python

get_table=pd.read_html("https://www.studyiq.com/articles/per-capita-income-of-india")
```

```python
data=get_table[1]
data
```

```python
df=data.drop([0])
df
```

```python
df.columns=['State/ UT','Per Capita Income (INR) 2020-21','Per Capita Income ($)2020-21']
print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
india_states=json.load(open('states_india.geojson','r'))
india_states['features'][0]
```

```python
india_states['features'][0].keys()
```

```python
state_id={}
for feature in india_states['features']:
    feature['id']=feature['properties']['state_code']
    state_id[feature['properties']['st_nm']]=feature['id']
    
```

```python
state_id
```

```python
#
#df.head()
df['id']=df['State/ UT'].apply(lambda x: state_id.get(x,7))
df
```

```python
df.drop(index=[18],axis=0,inplace=True)
```

```python
df.at[26,'id']=1
df.at[13,'id']=12
df

#df.drop(index=[17],axis=0,inplace=True)
```

```python
fig=px.choropleth(df,
             locations='id',
             geojson='india_states',
             color='Per Capita Income (INR) 2020-21',
             hover_name='State/ UT',
             
             )
fig.update_goes(fitbounds='locations',visible=False)
fig.show(renderers='google-chrome')
```
